```python
def introduction(): #function does not have parameters and values. it will only print what the program does and how to end the set of data
  print("Welcome!")
  print("This program will ask for three integers and then calculate the average.")
  print("It will also determine if the integers are equal to, greater than, and less than the average.")
  print("If you would like to end the program, please enter 'n' when entering your first integer.") #special combination that skips to step 5 (print number of data sets)
  print("Once you are done, the program will print your integers, the average, how your integers compare to the average, and the number of sets you have entered.")
  print("Thank you!")
  print("Created by Jeffery Liu")
```

```python
def findaverage(num1, num2, num3): #function to determine and return the average of the integers (will total the integers and divide by the number of integers). receives the parameters: num1, num2, and num3
  average = ((num1 + num2 + num3)/3)
  return average
```

```python
def comparetoaverage(num1, num2, num3, average): #function to print and compare whether integers are equal to, greater than, and less than average (using if conditions)
#as well as printing the number of times an integer is equal to the average using a counter. receives the parameters: num1, num2, num3, and average (from findaverage function)

  isequal = 0 #counter to count the number of times an integer is equal to the average

  if num1 == average:
    print(f"{num1} is equal to the average.")
    isequal += 1 #adds 1 each time it is true
  elif num1 > average:
    print(f"{num1} is above the average.")
  else:
    print(f"{num1} is below the average.")
  if num2 == average:
    print(f"{num2} is equal to the average.")
    isequal += 1
  elif num2 > average:
    print(f"{num2} is above the average.")
  else:
    print(f"{num2} is below the average.")
  if num3 == average:
    print(f"{num3} is equal to the average.")
    isequal += 1
  elif num3 > average:
    print(f"{num3} is above the average.")
  else:
    print(f"{num3} is below the average.")

  print(f"{isequal} value or values equal to the average.")
```

```python
def main(): #main function that calls and ties in previously created functions to print. it has a loop that has a special combination input to determine if user wants to stop or continue.
#it will ask for three integers: num1, num2, and num3 that will act as parameters for the previous functions. it also has counter that will determine and print the number of times sets of data were entered. the function itself has no parameters.
  introduction()

  datasets = 0 #counter to count the number of times sets of three data values were entered

  while True:
      num1 = (input("\nEnter an integer ('n' to stop): "))
      if num1 == 'n': #this is the special combination and it is used to determine if user wants to continue or stop (once stopped, it will print the number of data sets entered)
        break
      num1 = int(num1) #since the first input can be a str, int, or float, i have to convert the input to int
      num2 = int(input("Enter another integer: "))
      num3 = int(input("Enter one more integer: "))

      print(f"\nThe three original integers are {num1} {num2} {num3}")

      average = findaverage(num1, num2, num3) #calls the function of findaverage to the variable 'average' so can be used to print and used for the next function, comparetoaverage
      print (f"The average is {average:.3f}")

      comparetoaverage(num1, num2, num3, average)

      datasets += 1 #adds 1 each time a loop is completed to show how many data sets were entered

  print(f"{datasets} sets of three data values were entered and processed.")
```

```python
main()
```
